```python
from sympy import *
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
var('x y u')
var('t', real=True)
```

```python
f = Lambda(x, (x**3 - 3*x + 2)*exp(-x/4) - 1)
g = Lambda(x, (2 - (x ** 2 - 1) ** .5))
h = Lambda(x, (abs((x-1)**.5)-2*sin(x)))
```

```python
print(f(0))
print(f(1))
print(f(2))
```

```python
plot(f(x))
```

```python
plot(g(x))
```

```python
# reta tangente
fl = Lambda(x, diff(f(x),x))
x0 = -1/2
r = Lambda(x, fl(x0)*(x-x0) + f(x0))
```

```python
plot(h(x))
```

```python
p1 = plot(h(x))
p2 = plot(r(x), (x,-1.5,1), line_color='red', show=False)

p1.extend(p2)

p1.show()
```
